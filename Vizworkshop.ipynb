```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
file_path1 = '/Users/seminmahn/Downloads/air_quality_health_impact_data.csv'
file_path2 = '/Users/seminmahn/Downloads/air_quality_health_impact_data 2.csv'
file_path3 = '/Users/seminmahn/Downloads/GlobalLandTemperaturesByCountry.csv'
air_quality_1 = pd.read_csv(file_path1)
air_quality_2 = pd.read_csv(file_path2)
global_temps = pd.read_csv(file_path3)

# Convert date column to datetime for the global temperatures dataset
global_temps['dt'] = pd.to_datetime(global_temps['dt'])

# Distribution plot of AQI from the first dataset
plt.figure(figsize=(10, 6))
sns.histplot(air_quality_1['AQI'], bins=30, kde=True)
plt.title('Distribution of Air Quality Index (AQI)')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()

# Scatterplot of AQI vs Respiratory Cases
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AQI', y='RespiratoryCases', data=air_quality_1, alpha=0.5)
plt.title('Scatterplot of AQI vs Respiratory Cases')
plt.xlabel('AQI')
plt.ylabel('Respiratory Cases')
plt.show()

# Heatmap showing correlations in the first air quality dataset
plt.figure(figsize=(12, 8))
correlation_matrix = air_quality_1.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Matrix of Air Quality Data')
plt.show()

# Global temperature trends over time
plt.figure(figsize=(14, 7))
sns.lineplot(data=global_temps, x='dt', y='AverageTemperature', ci=None, alpha=0.5)
plt.title('Global Average Temperature Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.show()

# Temperature trends for the United States
us_temps = global_temps[global_temps['Country'] == 'United States']
plt.figure(figsize=(14, 7))
sns.lineplot(data=us_temps, x='dt', y='AverageTemperature', ci=None, color='orange')
plt.title('Temperature Trends in the United States Over Time')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.show()
```
